```python
import numpy as np
import pandas as pd
import altair as alt
unemploymentData = pd.read_csv("UNRATE.csv")
dogeData = pd.read_csv("DOGE-USD.csv")
```

```python
dogeData
```

```python
unemploymentData
```

```python
#unemploymentData
unemploymentChart = alt.Chart(unemploymentData).mark_area().encode(
    x=alt.X("DATE",title="Date",timeUnit="yearmonth"),
    y="UNRATE",
    tooltip=["yearmonth(DATE)","UNRATE"])
```

```python
#dogeData
dogeChart = alt.Chart(dogeData).mark_area().encode(
    x=alt.X("Date",title="Date",timeUnit="monthdate"),
    y="Close",
    tooltip=["yearmonth(Date)","Close"]).properties(width=900,height=300)
dogeChart
```

```python
alt.vconcat(unemploymentChart,dogeChart)
```
