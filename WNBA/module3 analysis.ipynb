```python
import numpy as np
import pandas as pd
import altair as alt

data = pd.read_csv("wnba-player-stats.csv")
pd.set_option('display.max_columns', 500)
```

```python
data # 3883 player entries, includes their stats for each season, from 1997 thru 2019.
```

```python
data1 = pd.read_csv("wnba-team-elo-ratings.csv")
data1 # team data for each game, from 1997 thru 2019
```

```python
alt.Chart(data).mark_line().encode(
    x="season",
    y=np.mean("prob1"),
    color="team1")
```

```python
topTenPER = data.groupby("Player").agg({"PER":np.mean}).sort_values("PER",ascending=False).head(10)
```

```python
topTenWinsGenerated = data.groupby("Player").agg({"Wins_Generated":np.mean}).sort_values("Wins_Generated",ascending=False).head(10)
topTenWinsGenerated
```

```python
yearsPlayed = lambda x:max(x)-min(x)
data.groupby("Player").agg({"Wins_Generated":sum,
                                   "year_ID":[min,max,yearsPlayed]}).sort_values([("Wins_Generated","sum")],ascending=False)
```

```python
# i want to make a chart that shows the trends of the top players in X statistical category, over the years. So, I'll
# start with Wins_Generated in order to give a baseline.
# Among the top 15 average Wins_Generated, plot the year on x axis and Wins_Generated on y axis. Each line is a player
```

```python
topTenWinsGenerated = data.groupby("Player").agg({"Wins_Generated":np.mean}).sort_values("Wins_Generated",ascending=False).iloc[0:10]

topWinnersRecords = data.loc[data["Player"].isin(topTenWinsGenerated.index)]
topWinnersRecords # now, we have every player who appears in the Top 15 Wins Generated (by average)
```

```python
alt.Chart(topWinnersRecords).mark_line().encode(
    x="year_ID", 
    y="Wins_Generated",
    color=alt.Color("Player",scale=alt.Scale(scheme="category10")),
    tooltip=["Player","Wins_Generated"]).properties(
    width=800,
    height=350)
```

```python
cols = ["Wins_Generated","PER","USG_pct","TS_pct"]

select_box = alt.binding_select(options=cols, name='Y-Axis')
sel = alt.selection_single(fields=['col'], bind=select_box, init={'col': 'Wins_Generated'})

alt.Chart(topWinnersRecords).transform_fold(
    cols,
    as_=["col","Y-Axis"]
).transform_filter(
    sel
).mark_line().encode(
    x="year_ID", 
    y="Y-Axis:Q",
    tooltip=["Player","year_ID","Y-Axis:Q"],
    color=alt.Color("Player",scale=alt.Scale(scheme="category10"))).properties(
    width=800,
    height=350).add_selection(sel)
```
